```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img1=mpimg.imread("7. Save Figure\RB.jpg")
img1
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img1)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img1)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img1,cmap="Accent")
plt.colorbar()
plt.show()
```
