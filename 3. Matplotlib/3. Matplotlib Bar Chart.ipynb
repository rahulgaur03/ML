```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python","Java","C","PHP","JS"]
class1_student=[30,23,34,67,54]
```

```python
plt.bar(classes,class1_student)
plt.show()
```

```python
plt.barh(classes,class1_student)
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.4,align='edge',color='r',edgecolor='g',linewidth=4)
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.4,align='edge',color='r',edgecolor='b',linewidth=4,alpha=0.2,linestyle=':')
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.4,align='edge',color='b',edgecolor='r',linewidth=4,label="Class 1 Student") #Use visible=False if you want to hide bar
plt.legend()
plt.show()
```
