```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python","JS","JAVA","C++","PHP"]
class1_student =[45,15,35,25,30]
```

```python
explode =[0.08,0,0.07,0,0] # To slice the piechart elements
color = ['b','r','y','c','g']
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",pctdistance=0.5,labeldistance=1.5)
plt.show()
```

```python
wedgeprops = {"linewidth":4,'width':1,'edgecolor':'k'}
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",counterclock=False,wedgeprops=wedgeprops,center=(2,3),frame=True)
plt.show()
```

```python
wedgeprops = {"linewidth":4,'width':1,'edgecolor':'k'}
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",counterclock=False,wedgeprops=wedgeprops,center=(2,3),frame=True,rotatelabels=True)
plt.show()
```

```python
wedgeprops = {"linewidth":4,'width':3,'edgecolor':'k'}
plt.figure(figsize=(12,12))
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",counterclock=False,wedgeprops=wedgeprops,center=(2,3),frame=True,rotatelabels=True)
plt.show()
```

```python
wedgeprops = {"linewidth":4,'width':1,'edgecolor':'k'}
plt.figure(figsize=(12,12))
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",counterclock=False,wedgeprops=wedgeprops,center=(2,3),frame=True)
plt.legend()
plt.show()
```
