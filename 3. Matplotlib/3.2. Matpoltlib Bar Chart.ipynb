```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python","Java","C","PHP","JS"]
class1_student=[30,23,34,67,54]
class2_student=[34,88,45,27,78]
class3_student=[21,65,83,69,41]
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.6,align='edge',color='k',edgecolor='r',linewidth=4,label="Class 1 Student") #Use visible=False if you want to hide bar
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.6,align='edge',color='k',edgecolor='r',linewidth=4,label="Class 1 Student")
plt.title("Bar Chart of Language Classes",fontsize=15)
plt.ylabel("Students")
plt.xlabel("Languages")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.2,color='k',linewidth=4,label="Class 1 Student")
plt.bar(classes,class2_student,width=0.2,color='r',linewidth=4,label="Class 2 Student")
plt.bar(classes,class3_student,width=0.2,color='b',linewidth=4,label="Class 3 Student")
plt.title("Bar Chart of Language Classes",fontsize=15)
plt.ylabel("Students")
plt.xlabel("Languages")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width =0.2
plt.bar(classes_index,class1_student,width,color='k',linewidth=4,label="Class 1 Student")
plt.bar(classes_index+width,class2_student,width,color='r',linewidth=4,label="Class 2 Student")
plt.bar(classes_index+width+width,class3_student,width,color='b',linewidth=4,label="Class 3 Student")
plt.xticks(classes_index+width,classes)
plt.title("Bar Chart of Language Classes",fontsize=15)
plt.ylabel("Students")
plt.xlabel("Languages")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width =0.2
plt.barh(classes_index,class1_student,width,color='k',linewidth=4,label="Class 1 Student")
plt.barh(classes_index+width,class2_student,width,color='r',linewidth=4,label="Class 2 Student")
plt.barh(classes_index+width+width,class3_student,width,color='b',linewidth=4,label="Class 3 Student")
plt.yticks(classes_index+width,classes)
plt.title("Bar Chart of Language Classes",fontsize=15)
plt.xlabel("Students")
plt.ylabel("Languages")
plt.legend()
plt.show()
```
