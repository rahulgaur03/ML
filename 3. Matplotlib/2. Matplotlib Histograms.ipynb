```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))
ml_student_age
```

```python
py_student_age = np.random.randint(15,50,(100))
py_student_age
```

```python
plt.hist(ml_student_age)
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8)
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.show()
```

```python

plt.hist(ml_student_age,bins,rwidth=0.8,orientation='horizontal')
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.show()
```

```python

plt.hist(ml_student_age,bins,rwidth=0.8,color='r')
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.show()
```

```python

plt.hist(ml_student_age,bins,rwidth=0.8,color='g',label="ML Student")
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python

plt.hist(ml_student_age,bins,rwidth=0.8,color='g',label="ML Student")
plt.hist(py_student_age,bins,rwidth=0.8,color='m',label="PY Student")
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python

plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,color=['b','r'],label=["ML Student","PY Student"])
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,color=['b','r'],label=["ML Student","PY Student"])
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python

style.use('ggplot')
plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,color=['b','r'],label=["ML Student","PY Student"])
plt.title("ML Students age histograms")
plt.xlabel("Student age category")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```
