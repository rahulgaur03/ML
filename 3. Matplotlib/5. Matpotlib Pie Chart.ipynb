```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python","JS","JAVA","C++","PHP"]
class1_student =[45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_student)
plt.show()
```

```python
explode =[0.08,0,0.07,0,0] # To slice the piechart elements
color = ['b','r','y','c','g']
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",shadow=True,radius=2,startangle=270)
plt.show()
```

```python
textprops={"fontsize":15}
plt.pie(class1_student,labels=classes,explode=explode,colors=color,autopct="%0.2f%%",radius=2,textprops=textprops)
plt.show()
```
