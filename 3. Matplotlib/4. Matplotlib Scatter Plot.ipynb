```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("4. Scatter Plot\googleplaystore.csv")
df.shape
```

```python
df = pd.read_csv("4. Scatter Plot\googleplaystore.csv",nrows=1000)
df
```

```python
x=df['Rating']
y=df['Reviews']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',marker='*',s=100)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',marker='*',s=100)
plt.scatter(x,df['Installs'],c='b',marker='o',s=80)
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews & Installs")
plt.show()
```
