```python
import pandas as pd
```

```python
sr1 = pd.Series([1,2,3])
sr1
```

```python
sr2 = pd.Series([4,5,6,7,8])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],"Name":["Madan","Mohan","Shyam","Raam"],"Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"ID":[5,6,7,8],"Name":["Manoj","Ratan","Ruchi","Ritika"],"Class":[10,9,12,11]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
df3=pd.DataFrame({"ID":[9,10],"Name":["Rajat","Vishal"],"Class":[12,11]})
df3
```

```python
pd.concat([df2,df3], axis=1)
```

```python
pd.concat([df2,df3], axis=1,join="inner")
```

```python
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
pd.concat([df1,df2],keys=['First','Second'])
```
