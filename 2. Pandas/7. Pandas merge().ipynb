```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"ID":[1,2,3,4,5],"Class":[9,10,11,12,8]})
df1
```

```python
df2 = pd.DataFrame({"ID":[1,2,3,4],"Name":["Rohan","Sohan","Amit","Sumit"]})
df2
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df2,df1,on='ID')
```

```python
pd.merge(df1,df2,on="ID",how="left") # How is work like joins in SQL
```

```python
pd.merge(df1,df2,on="ID",how="right")
```

```python
pd.merge(df1,df2,on="ID",how="outer")
```

```python
pd.merge(df1,df2,on="ID",how="outer",indicator=True)
```

```python
df3 = pd.DataFrame({"ID":[5,6,7,8],"Section":["A","B","C","D"]})
df3
```

```python
pd.merge(df2,df3,left_index=True,right_index=True)
```

```python
pd.merge(df2,df3)
```

```python
df4 = pd.DataFrame({"ID":[5,6,7,8],"Section":["A","B","C","D"]})
df4
```

```python
pd.merge(df3,df4,on="ID")
```

```python
pd.merge(df3,df4,on="ID",suffixes=('_First','_Second'))
```
