```python
import pandas as pd
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],"Name":["Madan","Mohan","Shyam","Raam"],"Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"Section":["A","B","C","D"],"Marks":[78,89,56,67]})
df2
```

```python
df1.join(df2) #Index must be same for join the Data Frames
```

```python
df3=pd.DataFrame({"Section":["A","B","C"],"Marks":[78,89,56]})
df3
```

```python
df1.join(df3) #If both have diffrent no. of value then it will show NaN
```

```python
df1.join(df3,how='right') # It will also work like join in SQL
```

```python
df1.join(df3,how='outer')
```

```python
df4=pd.DataFrame({"ID":[5,6,7,8],"Name":["Manoj","Ratan","Ruchi","Ritika"],"Class":[10,9,12,11]})
df4
```

```python
df1.join(df4,lsuffix="_left")
```
