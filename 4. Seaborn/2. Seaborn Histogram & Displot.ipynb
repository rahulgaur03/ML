```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
tip_df = sns.load_dataset('tips')
tip_df
```

```python
sns.distplot(tip_df['size'])
plt.show()
```

```python
sns.distplot(tip_df['tip'])
plt.show()
```

```python
sns.distplot(tip_df['total_bill'])
plt.show()
```

```python
sns.distplot(tip_df['total_bill'],bins=100)
plt.show()
```

```python
sns.distplot(tip_df['total_bill'],bins=100,hist=False)
plt.show()
```

```python
sns.displot(tip_df['total_bill'],kde=True) # distplot == displot
plt.show()
```

```python
sns.displot(tip_df['total_bill'],kde=True,rug=True)
plt.show()
```

```python
sns.distplot(tip_df['total_bill'],fit=norm,kde=False,color='r')
plt.show()
```

```python
sns.distplot(tip_df['total_bill'],fit=norm,kde=False,color='r',vertical=True)
plt.show()
```

```python
sns.displot(tip_df['total_bill'],label="Total Bill")
plt.legend()
plt.show()
```
