```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips') #https://github.com/mwaskom/seaborn-data
tips
```

```python
sns.lineplot(x='size',y='total_bill',data=tips,hue='sex',style='sex',palette='hot')
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=tips,hue='sex',style='sex',dashes=False,markers=['o','<'])
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(16,9))
sns.lineplot(x='size',y='total_bill',data=tips,hue='sex',style='sex',dashes=False,markers=['o','<'])
plt.title(" Member Bill Ratio")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(x='size',y='total_bill',data=tips,hue='day',style='day',dashes=False,markers=['o','<','>','^'])
plt.title(" Member Bill Ratio")
plt.show()
```
